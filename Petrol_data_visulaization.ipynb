```python
plt.plot(df['Year'], df['USA'], marker = "*", label = "Canada")
plt.plot(df['Year'], df['Japan'], marker = ".", label = "Japan")
plt.plot(df['Year'], df['Australia'], marker = "o", label = "Australia")
plt.title("Pertrol prices over time in different countries", fontdict = {'fontname': 'sans-serif', 'fontsize': 20, 'color': '#d08d88'})
plt.xlabel("Years", fontdict = {'fontname': 'sans-serif', 'fontsize': 20, 'color': '#d08d88'}, labelpad = 15)
plt.ylabel("Price in dollars", fontdict = {'fontname': 'sans-serif', 'fontsize': 20, 'color': '#d08d88'}, labelpad = 15)
plt.legend()

```

```python
plt.figure(figsize=(10,5)) # (height, width)

for i in df.columns[1:]:
  plt.plot(df['Year'], df[i], label = i, marker = '*')
plt.legend()
```
